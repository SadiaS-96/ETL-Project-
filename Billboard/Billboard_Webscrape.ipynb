```python
from bs4 import BeautifulSoup
import requests
import os
import pymongo
import pandas as pd
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.billboard_db
collection = db.items
```

```python
filepath = os.path.join("Billboard_top_100.html")
with open(filepath, encoding='utf-8') as file:
    html = file.read()
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

```python
#Find the top 200 best songs
top_200 = soup.find("ol", class_='chart-list__elements').\
    find('span', class_='chart-element__information__song').text

#number_subscribers = soup.find("span", class_='subscribers').\
 #   find('span', class_='number').text
#print(f"The number of subscribers: {number_subscribers}")

print(top_200)
```

```python
#Find the top 200 best songs
top_200 = soup.find("ol", class_='chart-list__elements').\
    find('span', class_='chart-element__information__song').text

#number_subscribers = soup.find("span", class_='subscribers').\
 #   find('span', class_='number').text
#print(f"The number of subscribers: {number_subscribers}")

print(top_200)
```

```python
#Find the top 200 best songs
top_200 = soup.find_all("span", class_='chart-element__information')





all_songs = []
print(len(top_200))
##print(top_200)
for songs in top_200:
    all_songs.append(songs.find('span', class_='chart-element__information__song').text)



from pprint import pprint
pprint(all_songs)

#number_subscribers = soup.find("span", class_='subscribers').\
 #   find('span', class_='number').text
#print(f"The number of subscribers: {number_subscribers}")

##print(top_200)
```

```python
all_artists = []

for artists in top_200:
    all_artists.append(artists.find('span', class_='chart-element__information__artist').text)



from pprint import pprint
pprint(all_artists)
```

```python
data = pd.DataFrame(list(zip(all_artists,all_songs)),columns=['Artist','Song'])
data.index += 1
data
                                                              
                                                              
```
