```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import os
```

```python
filepath = os.path.join("apple_top_100.html")
with open(filepath, encoding='utf-8') as file:
    html = file.read()
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
results = soup.find_all("div", class_ = "tracklist-item__text")

artist_list = []
for result in results:
   
    artist = result.find('div',class_="we-truncate").a.text
    artist_list.append(artist)
    
artist_list
```

```python
#songs

results = soup.find_all("a", class_ = "tracklist-item__text__link")

song_list = []
for result in results:
   
    song = result.find('div',class_="spread").span.text
    song_list.append(song)
    
song_list
```

```python
df = pd.DataFrame(list(zip(song_list, artist_list)), columns=['Track', 'Artist'])
df.index +=1
df
```

```python
df.to_csv("apple_top_100.csv")
```
